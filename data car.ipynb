```python
#summary of year
a=df['year'].mean()
print("SMA of year=",np.floor(a))
print("Most frequent value of year=",df['year'].mode())
#summary of selling_price
print("SMA of selling_price=",df['selling_price'].mean())
print("Most frequent value of selling_price=",df['selling_price'].mode())
print("max of selling_price=",df['selling_price'].max())
print("min of selling_price=",df['selling_price'].min())
#summary of km_driven
print("SMA of  km_driven=",df['km_driven'].mean())
print("Most frequent value of km_driven=",df['km_driven'].mode())
print("max of  km_driven=",df['km_driven'].max())
print("min of  km_driven=",df['km_driven'].min())
#summary of fuel
print("Most frequent value of fuel=",df['fuel'].mode())
#summary of seller_type
print("Most frequent value of seller_type=",df['seller_type'].mode())
#summary of transmission
print("Most frequent value of transmission=",df['transmission'].mode())
#summary of owner
print("Most frequent value of owner=",df['owner'].mode())
```
